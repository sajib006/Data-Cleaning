```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("banking.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["campaign"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss= StandardScaler()
ss.fit(dataset[["campaign"]])
```

```python
dataset["cmpaign_ss"]=pd.DataFrame(ss.transform(dataset[["campaign"]]),columns=["x"])
```
