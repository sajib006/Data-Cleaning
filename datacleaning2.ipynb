```python
dt = pd.DataFrame({"size":["s","m","l","s","s","l"]})
dt.head(3)
```

```python
ord_data = [["s","m","l"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(dt[["size"]])
```

```python
dt["size_en"]=oe.transform(dt[["size"]])
```

```python
dt
```

```python
ord_data1= {"s":5,"m":6,"l":2}
```

```python
dt["size_en"]=dt["size"].map(ord_data1)
```
